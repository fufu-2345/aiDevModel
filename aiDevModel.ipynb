```python
from transformers import VitsModel, AutoTokenizer
import torch
import scipy.io.wavfile as wavfile
```

```python
with open("sian.txt", "r", encoding="utf-8") as f:
    text = f.read().strip()
inputs = tokenizer(text, return_tensors="pt")
```

```python
model_name = "VIZINTZOR/MMS-TTS-THAI-MALE-NARRATOR"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = VitsModel.from_pretrained(model_name)

with torch.no_grad():
    audio = model(**inputs).waveform[0].numpy()

wavfile.write("VIZINTZOR/MMS-TTS-THAI-MALE-NARRATOR.wav", model.config.sampling_rate, audio)
```
